## Figurtall
I matematikken snakker vi gjerne om figurtall. Vi kaller dem det fordi de kan representeres geometrisk. Nedenfor ser du eksempler på henholdsvis trekanttall og kvadrattall.

```python
import matplotlib.pyplot as plt

def draw_triangle_numbers(n):
    fig, axes = plt.subplots(1, n, figsize=(n * 3, 3))
    if n == 1:
        axes = [axes]  # Sørger for at axes alltid er en liste
    
    for i in range(n):
        axes[i].set_xlim(-i-1, i + 1)
        axes[i].set_ylim(-i-1, i + 1)
        axes[i].set_xticks([])
        axes[i].set_yticks([])
        axes[i].set_title(f'T_{i+1} = {(i+1)*(i+2)//2}')
        axes[i].set_frame_on(False)  # Fjerner rammen rundt subplottene
     
        count = 0
        for row in range(i + 1):
            for col in range(row + 1):
                axes[i].scatter(col, -row, s=200, color='blue')
                count += 1
                
    plt.show()

# Tegner de fire første trekanttallene
draw_triangle_numbers(4)
```

### Trekanttall
La oss begynne med trekanttallene.
Vi ser at verdien av det første trekanttallet er lik 1.

$a_1 = 1$

Videre ser vi at vi får det andre trekanttallet ved å legge til 2, og det tredje ved å legge til 3, og så videre.

$a_2 = a_1 + 2$

$a_3 = a_2 + 3$

Ved å sette $a_0=0$ så kan vi si at trekanttallene følger følgende formel:

$a_n = a_{n-1} + n$


La oss lage et program som regner ut de 10 første trekanttallene, og skriver dem ut til konsoll.

```python
trekanttall = 0
for i in range(1, 11,1):
    trekanttall = trekanttall + i
    print(trekanttall)
```
